```python
import os
os.system("pip install compmec-nurbs")
from compmec import nurbs
import numpy as np
from matplotlib import pyplot as plt
```

```python
degree, npts = 1, 5
knotvector = nurbs.GeneratorKnotVector.uniform(degree, npts)
print("knot vector = ", knotvector)
```

```python
N = nurbs.SplineBaseFunction(knotvector)
dN = nurbs.SplineBaseFunction(knotvector)
dN.derivate()
uplot = np.linspace(0, 1, 1025)

fig, axis = plt.subplots(1, npts, figsize=(20, 6))
for i in range(npts):
    axis[i].plot(uplot, N[i, degree](uplot), label=r"$N_{%d%d}$"%(i,degree))
    axis[i].legend()
    axis[i].grid()
fig.suptitle(f"Base functions of degree {degree}")

fig, axis = plt.subplots(1, npts, figsize=(20, 6))
for i in range(npts):
    axis[i].plot(uplot, dN[i, degree-1](uplot), label=r"$N_{%d%d}'$"%(i,degree))
    axis[i].legend()
    axis[i].grid()
fig.suptitle(f"Derivative of base functions of degree {degree}")
```
