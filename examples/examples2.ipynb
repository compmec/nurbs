```python
import os
os.system("pip install compmec-nurbs")
from compmec import nurbs
import numpy as np
from matplotlib import pyplot as plt
```

```python
degree, npts = 2, 7
knotvector = nurbs.GeneratorKnotVector.uniform(degree, npts)
print("knot vector = ", knotvector)
```

```python
N = nurbs.SplineFunction(knotvector)
uplot = np.linspace(0, 1, 1025)
for j in range(degree+1):
    fig, axis = plt.subplots(1, npts, figsize=(20, 6))
    for i in range(npts):
        axis[i].plot(uplot, N[i, j](uplot), label=r"$N_{%d%d}$"%(i,j))
        axis[i].legend()
        axis[i].grid()
    fig.suptitle(f"Base functions of degree {j}")
```
